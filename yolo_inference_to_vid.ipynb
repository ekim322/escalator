```python
import os
import glob
import cv2

import matplotlib.pyplot as plt

from yolo_infer import predict_paths
```

```python
model_path = 'weights/best_ckpt.pt'
img_dir = 'esc_dynamic'
```

```python
pred_lst = predict_paths(model_path, img_dir)

imgs = [x[0] for x in pred_lst]
heights = [x[1] for x in pred_lst]
```

```python
def put_text(img, text):
    vid_h = img.shape[0]
    vid_w = img.shape[1]
    
    cv2.putText(img=img, text=text,org=(int(vid_w/20),int(vid_h/7)), 
                fontFace=cv2.FONT_HERSHEY_DUPLEX, fontScale=4, color=(255,0,0), thickness=7)
```

```python
vid_name = 'oh_man.avi'

vid_h = imgs[0].shape[0]
vid_w = imgs[0].shape[1]
vid_path = os.path.join(os.getcwd(), vid_name)

video = cv2.VideoWriter(vid_path, cv2.VideoWriter_fourcc(*'MJPG'), 15, (vid_w, vid_h))

# prev_height = heights[0]

for i, image in enumerate(imgs):
    if i==0:
        continue
    
    img = image.copy()
    cur_height = heights[i]
    
    if cur_height==heights[i-1] and cur_height==heights[i-2]: # Look at 2 previous (sometimes 2 consec bbox same size even if moving)
        put_text(img, 'Not Moving')
        print("Frame {}, Cur Height {} - Not Moving".format(i, cur_height))
    else:
        put_text(img, 'Moving')
        print("Frame {}, Cur Height {} - Moving".format(i, cur_height))

    video.write(img)
```
